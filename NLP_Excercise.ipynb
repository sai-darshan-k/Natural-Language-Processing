# Tokenization
Word and Sentence Tokenization
Tokenization in natural language processing (NLP) refers to the process of breaking down text into smaller units, typically words or sentences, known as tokens. These tokens are the basic building blocks of any NLP task and can serve as the foundation for further analysis.

Word tokenization involves splitting a piece of text into individual words or terms, while sentence tokenization involves splitting a piece of text into individual sentences. This process is crucial in NLP as it allows for easier analysis, manipulation, and understanding of textual data.


**The punkt resource is used for tokenization, specifically for word and sentence tokenization.
````python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

def find_word_frequency(text):
    # Tokenize the text into words
    words = word_tokenize(text)

    # Calculate frequency distribution
    frequency_distribution = FreqDist(words)

    return frequency_distribution

# Sample text
sample_text = "Natural language processing (NLP) is a subfield of linguistics, " \
              "computer science, and artificial intelligence concerned with " \
              "the interactions between computers and human language, " \
              "in particular how to program computers to process and analyze large " \
              "amounts of natural language data."

# Calculate word frequency
word_frequency = find_word_frequency(sample_text)

# Print the frequency of each word
print("Word\t\tFrequency")
for word, frequency in word_frequency.items():
    print(f"{word}\t\t{frequency}")
````

# WORD VECTOR REPRESENTATION
## The word vectorization or embedding is the process of converting text into numerical vectors in a vector space.
##Each numerical value corresponding to the words can preserve some semantic information about the word and that may be lexical or compositional semantics.

#*   Sparse or frequency-based vectors - The sparse vector representation captures less semantic information with exploiting considerable amount of memory.

## Eg. - Bag of Words (BoW) or One Hot Encoding, Term Frequency and Inverse Document Frequency (Tf-Idf) and Distributed representations are the some of the sparse representation methods.

#	Bag of Words or One Hot Encoding
#One hot encoding or Bag of Words (BoW) is a binary vector, which means only one index holds the non-zero value to represent the presence of word and the remaining indices will be zero.
#It is also termed as **count vectorization** and dummy variables.
#This method is very effective to represent categorical values.
#There are mainly two shortcomings associated with the one hot encoding.
#Initially, the curse of dimensionality problem, when the data with a higher dimension then the matrix taken up to huge amount of zeros other than only one index with 1 to represent the presence of word.
#This leads to the issue of loss of memory.
#The second one is this method is hard to extract the semantics.

#Max_df stands for maximum document frequency.
#Itignore words which occur frequently.
#These words could be like the word the that occur in every document and does not provide and valuable information to our text classification or any other machine learning model and can be safely ignored.
#Max_df looks at how many documents contain the word and if it exceeds the max_df threshold then it is eliminated from the sparse matrix.

````python
#The words "NLP, is, Computer" have been removed from the sparse matrix as they occur in more than 1 document.
from sklearn.feature_extraction.text import CountVectorizer

document = ["NLP is a field of Computer Science and Linguistics",
            "NLP is fun",
            "NLP makes humans computer interaction"]

# Create a Vectorizer Object
vectorizer = CountVectorizer(max_df=1)

vectorizer.fit(document)

# Printing the identified Unique words along with their indices
print("Vocabulary: ", vectorizer.vocabulary_)

# Encode the Document
vector = vectorizer.transform(document)

# Summarizing the Encoded Texts
print("Encoded Document is:")
print(vector.toarray())
````

# TF-IDF Vectorization


````python
_# import required module
from sklearn.feature_extraction.text import TfidfVectorizer

# assign documents
d0 = "Dogs are cute"
d1 = "Dogs vary of different breeds"
d2 = "Dogs are playful"

# merge documents into a single corpus
string = [d0, d1, d2]

# create object
tfidf = TfidfVectorizer()

# get tf-df values
result = tfidf.fit_transform(string)

# get idf values
print('\nidf values:')
for ele1, ele2 in zip(tfidf.get_feature_names_out(), tfidf.idf_):
    print(ele1, ':', ele2)

# get indexing
print('\nWord indexes:')
print(tfidf.vocabulary_)

# display tf-idf values
print('\ntf-idf value:')
print(result)
print('\n Type and Shape:',type(result), result.shape) #Compressed Sparse Row matrix

# in matrix form
print('\ntf-idf values in matrix form:')
print(result.toarray())
````

#Word2Vec ushered a new era in the history of word embedding, which is the popular technique utilizing shallow neural network for learning representations. Word2Vec is developed by Tomas Mikolov in 2013. Word2Vec is a combination of two neural network-based methods: Continuous Skip Gram and Continuous Bag of Words (CBoW).

#Continuous bag-of-words model: predicts the middle word based on surrounding context words. The context consists of a few words before and after the current (middle) word. This architecture is called a bag-of-words model as the order of words in the context is not important.
#Continuous skip-gram model: predicts words within a certain range before and after the current word in the same sentence.

#Skip-gram predicts the context words from the target word.
#CBoW predicts the target word based on the surrounding context or neighbour words.
#Skip Gram works well with small amount of data, while CBoW provides faster and efficient representations for frequent words.
#The major difference between Skip Gram and CBoW is in the target variable, there will be multiple one hot encoded target variables.

##Sentence embedding represents the sentences with similar meaning occupy the neighbouring vector space.

#	Doc2Vec
##The Doc2Vec is an unsupervised model working similar to the Word2Vec.
##This model incorporates two the paragraph vector models - PVDM (Distributed Memory version of Paragraph Vector) and PVDOBW (Distributed Bag of Words version of Paragraph Vector).
##PVDM predicts the next sentence in the context, which is similar to the CBoW and PVDOBW is similar to the Skip-gram in Word2Vec.

#	SentenceBERT
##SentenceBERT is a Transformer-BERT-based pre-trained sentence embedding mechanism with four key concepts: Attention, Transformers, BERT and Siamese Network.
##SentenceBERT model utilizes the potential of cosine similarity and Siamese network to extract the embedding from sentences.


#	GloVe-InferSent
##GloVe-InferSent is a supervised word embedding model pre-trained with Natural language Inference (NLI) data and SNLI (Stanford Natural Language Inference) dataset.
##The embedding created with the encoding of a pair of sentences. There are two versions of Infersent version1 uses GloVe and Infersent version 2 uses Fast-Text embedding models.


#	Universal Sentence Encoder [USE]
##Universal Sentence Encoder is one of the unusual embedding models and that employs the potential of attention mechanism.
##USE constitutes a pair of encoding models - Transformer and Deep Averaging Network (DAN).
##USE can utilizes either of two aforementioned encoding models. Initially, the model was undergone through the text pre-processing and later the model transforms the sentences to 512- dimensional vector.
##DAN estimates the N-grams and estimates the average of the embedding, later fed it into the Deep Neural Network (DNN) to get the sentence vector.

````python
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)

iteration =30000
for i in range(iteration):
    # input is X_train which is one hot encoded word
    # label is Y_train which is one hot encoded neighbor word
    sess.run(train_op, feed_dict={x: X_train, y_label: Y_train})
    if i % 3000 == 0:
        print('iteration '+str(i)+' loss is : ', sess.run(loss, feed_dict={x: X_train, y_label: Y_train}))
````

````python
from sentence_transformers import SentenceTransformer
#sbert_model = SentenceTransformer('bert-base-nli-mean-tokens')
model = SentenceTransformer('bert-base-nli-mean-tokens')
````

````python
sentences = [" JamesHutton is often viewed as the first modern geologist.  ",
             " In 1785 he presented a paper entitled Theory of the Earth to the Royal Society of Edinburgh.",
             " In his paper, he explained his theory that the Earth must be mucholder than had previously been supposed in order to allow enough time for mountains to be eroded and for sediments to form new rocks at the bottom of the sea, which in turn were raised up to become dry land.  ",
             " Huttonpublished a two-volume versionof his ideas in 1795 (Vol. 1, Vol. 2)"]
````

````python
sentence_embeddings = model.encode(sentences)
````

````python
query = "Who is viewed as the first Modern Geologist"
query_vec = model.encode([query])[0]
````

````python
def cosine(u, v):
    return np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))
````

````python
for sent in sentences:
  sim = cosine(query_vec, model.encode([sent])[0])
  print("similarity = ", sim, " \tSentence = ", sent )
````
